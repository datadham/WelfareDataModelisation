```python
diabetes = pd.read_csv('../data/diabetes_data.csv')
stroke = pd.read_csv('../data/stroke_data.csv')
hypertension = pd.read_csv('../data/hypertension_data.csv')
```

## General Association

Explore the links between "Age" and "BMI" to study the distribution of weight at different ages.

Compare "Sex" with other variables to identify differences between sexes in terms of health and behaviors.

- Classification	BMI range - kg/m2
- Severe Thinness	< 16
- Moderate Thinness	16 - 17
- Mild Thinness	17 - 18.5
- Normal	18.5 - 25
- Overweight	25 - 30
- Obese Class I	30 - 35
- Obese Class II	35 - 40
- Obese Class III	> 40

```python
diabetes.groupby('Sex')[["Age","BMI"]].agg(['mean','std','median','min','max','count']).T
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(diabetes['Age'], diabetes['BMI'], alpha=0.5)
plt.title('Age vs. BMI')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='BMI', data=diabetes)
plt.title('BMI Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('BMI')
plt.show()
```

```python
correlation_matrix = diabetes[['Age', 'BMI']].corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
diabetes[diabetes['Sex'] == 0]['Age'].hist(alpha=0.5, label='Male', bins=15)
diabetes[diabetes['Sex'] == 1]['Age'].hist(alpha=0.5, label='Female', bins=15)
plt.title('Age Distribution by Sex')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
diabetes[diabetes['Sex'] == 0]['BMI'].hist(alpha=0.5, label='Male', bins=15)
diabetes[diabetes['Sex'] == 1]['BMI'].hist(alpha=0.5, label='Female', bins=15)
plt.title('BMI Distribution by Sex')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```

## Summary

The analysis shows that the distributions of Age and BMI are very similar between the two groups. There are no significant differences in the mean, median, or range of Age and BMI values between the two groups. This indicates a consistent pattern across both sexes for the given dataset.
